This program calculates the height of a ball that is dropped from a tower of a specified height after a given time.

```
#no libraries to import
```

Inputs:  

height: height of tower ($m$)

time: time interval the ball has fallen for ($s$)

```
tower_height = 250.0 #meters
time = 1.0 #seconds
```

Constants:

g: gravitational acceleration ($m s^{-2}$)

```
g = 9.81
```

Calculate the distance the ball has dropped using the formula:

$ z = g * \frac{{time}^2}{2}$ 

Calculate its height by the formula

$ball\_height = tower\_height - z$

and output results.

```
z = g * time**2 /2
ball_height = tower_height - z
print("The height of the ball is:", ball_height,"meters")
```
