#C.

Modify your program further to calculate and plot the running average
  of the data, defined by

$
Y_k = {1\over2r} \sum_{m=-r}^r y_{k+m}\,,
$

where $r=5$ in this case (and the $y_k$ are the sunspot numbers).  Have the
program plot both the original data and the running average on the same
graph, again over the range covered by the first 1000 data points.

````
import numpy as np
import matplotlib.pyplot as plt
````

Set the variable r as the size of the running average window

````
r = 5
````

Define a function for the running average.  Take as input:

raw_data: the array with the elements

xval: the element to find the running average for

window: the width of the window

Use the variable:

vals: the array storing elements you want to average

x_min: the minimum element of the raw_data array you are going to splice

x_max: the maximum element of the 

aver: the average you want to return

Note that you'll want to pay special attention to the edge cases, that is when xval<window or when xval+window> length(raw_data).  How you handle this is somewhat up to you, here I'm just putting into the average anythign I can.

Also, there are a number of different ways you could program this.  For example, this could all be done with the single line of code:



```
return np.mean(raw_data(np.min([0,xval-window]):np.max([xval+window,raw_data.size]))
```

But of course that's kinda hard to read.  Can you unpack it to understand what it does?


````
def running_aver(raw_data,xval,window):
  """
  Calculates the running average for element xval in raw_data array with window width window
  """
  x_min = np.max([0,xval-window])
  x_max = np.min([xval+window,raw_data.size])
  vals = raw_data[x_min:x_max]
  aver = np.sum(vals) / vals.size
  return aver
````

First read the data into the arrays months and sunspots, then plot. Then create a new array called averages and use a for loop to fill it with the running averages of the data. Only plot the first 1000 points

````
months, sunspots = np.genfromtxt('http://www-personal.umich.edu/~mejn/cp/data/sunspots.txt',unpack=True)
average=np.zeros(sunspots.size)
for i in np.arange(sunspots.size):
  average[i] = running_aver(sunspots,i,r)
plt.plot(months[0:1000],sunspots[0:1000],'.')
plt.plot(months[0:1000],average[0:1000],'r')
plt.xlabel('Months since January 1749')
plt.ylabel('Number of sunspots')
plt.title('Number of sunspots since 1749')
plt.legend(['Raw Data','Running Average'])
plt.show()
````
