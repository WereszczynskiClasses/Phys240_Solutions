#Problem 2: Altitude of a satellite
A satellite is to be launched into a circular orbit around the Earth so that it orbits the planet once every $T$ seconds.  It can be shown that the altitude $h$ above the Earth's surface that the satellite orbits is:

$R = 6371 km$ is the Earth's radius.

# C 

Technically a geosynchronous satellite is one that orbits the Earth once per *sidereal day*, which is 23.93 hours, not 24 hours. Why is this? And how much difference will it make to the altitude of the satellite?

The 24 hour day accounts for the amount of time it takes for the earth to rotate once about its axis and also rotate a bit extra to place the sun (or moon) back at its original location.  But a geosynchronous satellite just needs to rotate about the earth's axis in a day. So we need to use the sidereal day.  The heights of the two  orbits are:

24 hour day: 35,855 km

23.93 day: 35,773 km

For a difference of 112 km, or ~.3% difference in height
