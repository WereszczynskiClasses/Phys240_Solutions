##A. 
Write two python functions.  The first should implement the function:

$
f\left(x\right) = \frac{3x}{0.4+ \left(x-2\right)^2}
$

The second should take a set of $k$ exact dataponts, $f\left(x\right)$, and a set of $k$ interpolated datapoints $g\left(x\right)$, and compute the root mean square error  $\sigma$ by:

$ \sigma = \sqrt{\frac{1}{N}\sum_{i=i}^k \left(f\left(x_i\right)-g\left(x_i\right)\right)^2}$

Define a function 

$
f\left(x\right) = \frac{3x}{0.4+ \left(x-2\right)^2}
$

```
def f(x):
  '''
  Input is one (or an array of) x-values
  Output is f(x)
  ''' 
  return 3*x/(0.4+(x-2)**2)
```

Define the root mean square error function:

$ \sigma = \sqrt{\frac{1}{N}\sum_{i=i}^k \left(f\left(x_i\right)-g\left(x_i\right)\right)^2}$

```
def rmse(f,g):
  '''
  Input is two arrays, f and g, each of which must be the same length
  Output is the root mean square error (which is a single float)
  '''
  return np.sqrt(np.mean((f-g)**2))
```

#B. 

Write a python program that uses the above functions to interpolate between the values of $f\left(x\right)$ sampled uniformly at $N=9$ points in the interval $0\leq x < 4$. 

Create a series of plots below where you compare the $N=9$ points to the following:

1.  The function $f$ computed with the above function.

2.  Nearest neighbor interpolation

3.  Linear interpolation

4.  Cubic spline interpolation

5.  Lagrange polynomial interpolation.

For each of these, compute estimated datapoints using a uniform spacing of $dx = 0.025$ over the range $0\leq x \leq 4$.  Be sure to use appropriate symbols and legends to identify different curves/points â€“ use multiple plots as needed for clarity.  For each interpolation use the function you wrote above to compute the RMSE.

Import the appropriate packages

```
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

User defined input: 

N: number of datapoints

dx: spacing between interpolated points x

x_min: minimum x value

x_max: maximum x value

```
N = 9
dx = 0.025
x_min = 0
x_max = 4
```

Make sure our functions from above are defined

```
def f(x):
  '''
  Input is one (or an array of) x-values
  Output is f(x)
  ''' 
  return 3*x/(0.4+(x-2)**2)

def rmse(f,g):
  '''
  Input is two arrays, f and g, each of which must be the same length
  Output is the root mean square error (which is a single float)
  '''
  return np.sqrt(np.mean((f-g)**2))
```

Compute $N$ datapoints at the points x_data and their corresponding values at y_data

Sample all interpolations at the values x_vals

Compute the exact answers as y_exact 

Note the different uses of linspace and arange below.  There are multiple ways you could make these arrays, but linspace is nice because it will give you the number of datapoints you request, and arange will space the datapoints out by a specified amount

```
x_data = np.linspace(x_min,x_max,N)
y_data = f(x_data)
x_vals = np.arange(x_min,x_max,dx)
y_exact = f(x_vals)
```

Compute and plot the nearest neighbor interpolation.  Compute the RMSE

```
f_nearest = interpolate.interp1d(x_data, y_data, kind='nearest') #function for nearest neighbor interpolation
y_nearest = f_nearest(x_vals)
plt.plot(x_data,y_data,'*')
plt.plot(x_vals,y_exact,'k')
plt.plot(x_vals,y_nearest,'r')
plt.title("Nearest Neighbor Interpolation")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
print("The RMSE for nearest neighbor is:",rmse(y_exact,y_nearest))
```

Compute and plot the linear interpolation.  Compute the RMSE

```
f_linear = interpolate.interp1d(x_data, y_data, kind='linear') #function for linear interpolation
y_linear = f_linear(x_vals)
plt.plot(x_data,y_data,'*')
plt.plot(x_vals,y_exact,'k')
plt.plot(x_vals,y_linear,'r')
plt.title("Linear Interpolation")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
print("The RMSE for linear is:",rmse(y_exact,y_linear))
```

Compute and plot the cubic spline interpolation.  Compute the RMSE

```
f_cubic = interpolate.interp1d(x_data, y_data, kind='cubic') #function for cubic spline interpolation
y_cubic = f_cubic(x_vals)
plt.plot(x_data,y_data,'*')
plt.plot(x_vals,y_exact,'k')
plt.plot(x_vals,y_cubic,'r')
plt.title("Cubic Interpolation")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
print("The RMSE for cubic is:",rmse(y_exact,y_cubic))
```

Compute and plot the Lagrange polynomical interpolation.  Compute the RMSE

```
p_lagrange = interpolate.lagrange(x_data,y_data)
y_lagrange = np.polyval(p_lagrange,x_vals)
plt.plot(x_data,y_data,'*')
plt.plot(x_vals,y_exact,'k')
plt.plot(x_vals,y_lagrange,'r')
plt.title("Lagrange Polynomial Interpolation")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
print("The RMSE for Lagrange polyomial is:",rmse(y_exact,y_nearest))
```
